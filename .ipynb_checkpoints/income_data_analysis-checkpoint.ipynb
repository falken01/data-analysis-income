```python
#Income dataset

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_income = pd.read_csv("https://community052.blob.core.windows.net/csvs/train.csv")
```

```python
#analiza podstawowa

df_income.groupby(['gender']).size().plot(kind='pie',ylabel="Podział na płeć", autopct='%1.0f%%')
df_income.groupby(['gender']).size()
```

```python
df_income.groupby(['race']).size().plot(kind='pie',ylabel="Podział na płeć", autopct='%1.0f%%')
df_income.groupby(['race']).size()
```

```python
numbers = []
age_ranges =["17-25 lat","26-35 lat","36-50 lat","51-68 lat","68+ lat"]
numbers.append(len(df_income[(df_income['age']>16) & (df_income['age']<=25)]))
numbers.append(len(df_income[(df_income['age']>25) & (df_income['age']<=35)]))
numbers.append(len(df_income[(df_income['age']>35) & (df_income['age']<=50)]))
numbers.append(len(df_income[(df_income['age']>50) & (df_income['age']<=68)]))
numbers.append(len(df_income[(df_income['age']>68)]))
```

```python
plt.bar(range(len(numbers)), numbers, width = 0.5, color='r')
plt.title('Podział grupy osób według wieku')
plt.ylabel('Liczba osób')
plt.xticks(np.arange(5), age_ranges)
plt.show()
```

```python
df_income.groupby(['income_>50K']).size().plot(kind='pie',ylabel="Czy osoba zarabia więcej niż $50k", autopct='%1.0f%%')
df_income.groupby(['income_>50K']).size()
```

```python
df_income[df_income["income_>50K"]>0].groupby(['race']).size().plot(kind='pie',ylabel="Podział na płeć", autopct='%1.0f%%')
```

```python
df_income[df_income["income_>50K"]>0].groupby(['gender']).size().plot(kind='pie',ylabel="Podział na płeć", autopct='%1.0f%%')
```

```python
df_income['income_>50K']>0

numbers = []
age_ranges =["17-25 lat","26-35 lat","36-50 lat","51-68 lat","68+ lat"]
numbers.append(len(df_income[(df_income['age']>16) & (df_income['age']<=25) & (df_income['income_>50K']>0)]))
numbers.append(len(df_income[(df_income['age']>25) & (df_income['age']<=35) & (df_income['income_>50K']>0)]))
numbers.append(len(df_income[(df_income['age']>35) & (df_income['age']<=50) & (df_income['income_>50K']>0)]))
numbers.append(len(df_income[(df_income['age']>50) & (df_income['age']<=68) & (df_income['income_>50K']>0)]))
numbers.append(len(df_income[(df_income['age']>68) & (df_income['income_>50K']>0)]))
```

```python
plt.bar(range(len(numbers)), numbers, width = 0.5, color='b')
plt.title('Podział grupy osób według wieku')
plt.ylabel('Liczba osób')
plt.xticks(np.arange(5), age_ranges)
plt.show()
```

```python
df_income['income_>50K']>0

numbers = []
age_ranges =["0-20 h/week","21-40 h/week","40-60 h/week","61+/week"]
hours = [0,20,40,60]
for i in range(0,len(hours),1):
    if len(hours) == i+1:
        mianownik = len(df_income[df_income['hours-per-week']>hours[i]])
        licznik = len(df_income[(df_income['hours-per-week']>hours[i]) & (df_income['income_>50K']>0)])
        result = licznik/mianownik
    else:
        mianownik = len(df_income[(df_income['hours-per-week']>hours[i]) & (df_income['hours-per-week']<=hours[i+1])])
        licznik = len(df_income[(df_income['hours-per-week']>hours[i]) & (df_income['hours-per-week']<=hours[i+1]) & (df_income['income_>50K']>0)])
        result = licznik/mianownik
    numbers.append(result)
```

```python
plt.plot(range(len(numbers)), numbers, marker="o", color='g')
plt.title('Podział osób najwięcej zarabiających według godzin pracy')
plt.ylabel("część wszsytkich osób")
plt.xticks(np.arange(4), age_ranges)
plt.show()
```
